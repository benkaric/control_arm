```python
import paho.mqtt.client as mqtt
import time
from Arm_Lib import Arm_Device
import json
import cv2
import ipywidgets.widgets as widgets
import threading
import enum



# 獲取機械臂對象
Arm = Arm_Device()
#宣告每個手臂關節的角度變數 初始值為90 即為置中



def on_connect(client, userdata, flags, rc):    #MQTT的連線function
    print("Connected with result code "+str(rc))

    # 將訂閱主題寫在on_connet中
    # 如果我們失去連線或重新連線時
    # 客戶端程式將會重新訂閱
    client.subscribe("NTUT/MQTT")
    
    

def on_message(client, userdata, msg):     #MQTT的接收訊息fUNCTION
    global androidJson,id
   
    # 轉換編碼utf-8才看得懂中文
    print(msg.topic+" " + msg.payload.decode('utf-8'))
    androidJson = json.loads(msg.payload.decode('utf-8'))
    sixid(msg)
    control(msg)
    if androidJson.keys() == {'mid'}:
        Arm.Arm_serial_servo_write6(80, 100, 70, 90, 140, 90, 500)
    
    
def sixid(msg): 
    global id,androidJson
    for i in androidJson.keys():
        if i == 'id1':
            id = androidJson['id1']
            print("切換位置:%d" % (id))
        if i == 'id2':
            id = androidJson['id2']
            print("切換位置:%d" % (id))
        if i == 'id3':
            id = androidJson['id3']
            print("切換位置:%d" % (id))
        if i == 'id4':
            id = androidJson['id4']
            print("切換位置:%d" % (id))
        if i == 'id5':
            id = androidJson['id5']
            print("切換位置:%d" % (id))
        if i == 'id6':
            id = androidJson['id6']
            print("切換位置:%d" % (id))
        if i == 'id7':
            id = androidJson['id7']
        return id
        
    

def rgb(msg):
    if msg.payload.decode('utf-8') == "紅":
        Arm.Arm_RGB_set(50, 0, 0) #RGB亮紅燈
        print("red")
    if msg.payload.decode('utf-8') == "綠":
        Arm.Arm_RGB_set(0, 50, 0) #RGB亮綠燈
        print("green")
    if msg.payload.decode('utf-8') == "藍":
        Arm.Arm_RGB_set(0, 0, 50) #RGB亮藍燈
        print("blue")
    
def control(msg):
    global id,androidJson
    if androidJson.keys() == {'angle'}:
        if id == 1:
            Arm.Arm_serial_servo_write(id, androidJson['angle'], 500)
        if id == 2:
            Arm.Arm_serial_servo_write(id, androidJson['angle'], 500)
        if id == 3:
            Arm.Arm_serial_servo_write(id, androidJson['angle'], 500)
        if id == 4:
            Arm.Arm_serial_servo_write(id, androidJson['angle'], 500)
        if id == 5:
            Arm.Arm_serial_servo_write(id, androidJson['angle'], 500)
        if id == 6:
            Arm.Arm_serial_servo_write(id, androidJson['angle'], 500)
            

            

# 連線設定
# 初始化地端程式
client = mqtt.Client()

# 設定連線的動作
client.on_connect = on_connect

# 設定接收訊息的動作
client.on_message = on_message

# 設定連線資訊(IP, Port, 連線時間)
client.connect("140.124.73.217", 1883, 60)

# 開始連線，執行設定的動作和處理重新連線問題
# 也可以手動使用其他loop函式來進行連接
client.loop_forever()

```
